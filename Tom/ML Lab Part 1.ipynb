```python
import pandas as pd
```

```python
orders = pd.read_csv("Orders.csv")
returns = pd.read_csv("Returns.csv")
```

```python
orders
```

```python
orders["Profit"] = orders["Profit"].replace('[\$,]', '', regex=True).astype(float)
```

```python
orders["Sales"] = orders["Sales"].replace('[\$,]', '', regex=True).astype(float)
```

```python
orders['Order.Date'] =  pd.to_datetime(orders['Order.Date'])
```

```python
orders['Ship.Date'] =  pd.to_datetime(orders['Ship.Date'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

SI = orders.groupby("Order.Date").Quantity.sum()
```

```python
plt.plot(SI)
plt.show()
```

```python
SIC = orders.groupby(["Order.Date","Category"]).Quantity.sum().reset_index()    
```

```python
import seaborn as sns
```

```python
returns.columns = ['Returned', 'Order.ID', 'Region']
```

```python
merge_df = pd.merge(orders, returns, on = "Order.ID", how = "left")
```

```python
merge_df['Year'] = merge_df['Order.Date'].dt.year
```

```python
return_loss = merge_df[~merge_df['Returned'].isnull()]
```

```python
import numpy as np

return_loss.groupby(['Year'])[['Profit']].agg(np.sum)
```
