```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,7
```

```python
orders = pd.read_csv('data/orders.csv')
returns = pd.read_csv('data/returns.csv')
```

```python
orders.head()
```

```python
orders.info()
```

```python
convert_to_float = lambda x: float(x.replace('$', '').replace(',', ''))
orders[['Profit', 'Sales']] = orders[['Profit', 'Sales']].applymap(convert_to_float)
orders.head()
```

```python
orders['Order.Date'] = pd.to_datetime(orders['Order.Date'])
orders['Ship.Date'] = pd.to_datetime(orders['Ship.Date'])
orders.head()
```

```python
_ = orders.groupby('Order.Date').Quantity.sum().plot()
```

```python
a = orders.groupby(['Order.Date', 'Category']).Quantity.sum().reset_index(level=1).groupby('Category')
for b, c in a:
     plt.plot(c)
```

```python
merged = orders.merge(returns, how='left', left_on='Order.ID', right_on='Order ID',)
merged.head()
```

```python
merged['year'] = merged['Order.Date'].dt.year
merged[~pd.isna(merged['Returned'])].groupby('year').Profit.sum().plot()
```

```python
returns_per_cust = merged[~pd.isna(merged['Returned'])].groupby('Customer.ID')['Order.ID'].nunique()

returns_per_cust[returns_per_cust > 1].count()

returns_per_cust[returns_per_cust > 5].count()
```

```python
returns_per_region = (merged.fillna('No').groupby(['Region_x', 'Returned'])['Order.ID'].nunique().
                                     reset_index(level=1).pivot(columns='Returned', values='Order.ID'))

returns_region_pct = returns_per_region['Yes'] / (returns_per_region['No'] + returns_per_region['Yes'])
returns_region_pct.sort_values(ascending=False)
```

```python
returns_per_subcat = (merged.fillna('No').groupby(['Sub.Category', 'Returned'])['Order.ID'].nunique().
                                     reset_index(level=1).pivot(columns='Returned', values='Order.ID'))

returns_subcat_pct = returns_per_subcat['Yes'] / (returns_per_subcat['No'] + returns_per_subcat['Yes'])
returns_subcat_pct.sort_values(ascending=False)
```
