```python
import numpy as np
import pandas as pd
```

```python
orders = pd.read_csv('Orders.csv')
returns = pd.read_csv('Returns.csv')
```

```python
orders.Profit = orders.Profit.str.replace('$','')
orders.Profit = orders.Profit.str.replace(',','')
orders.Sales = orders.Sales.str.replace('$','')
orders.Sales = orders.Sales.str.replace(',','')
```

```python
orders.Profit = orders.Profit.astype(float)
orders.Sales = orders.Sales.astype(float)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
#sns.set()

from pylab import rcParams
rcParams['figure.figsize'] = 15, 7
```

```python
orders['Order.Date'] = pd.to_datetime(orders['Order.Date'])
orders['Ship.Date'] = pd.to_datetime(orders['Ship.Date'])
```

```python
fig1_data = orders.groupby('Order.Date').Quantity.sum()
plt.plot(fig1_data)
plt.show()
```

```python
fig2_data = orders.groupby(['Order.Date','Category']).Quantity.sum().reset_index()
```

```python
sns.lineplot(x="Order.Date", y='Quantity', hue = 'Category', data=fig2_data)
```

```python
sns.FacetGrid(fig2_data, col = "Category", hue="Category").map(plt.plot, 'Order.Date', 'Quantity').add_legend()
```

```python
df = orders.merge(returns, how='left', on=None, left_on='Order.ID', right_on='Order ID')
```

```python
df['Year'] = df['Order.Date'].dt.year
```

```python
return_loss = df[~df['Returned'].isnull()]
return_loss = return_loss.groupby('Year')
```

```python
region_orders = df.groupby('Region_x').nunique()['Order.ID'].reset_index()
region_returns = df.groupby(['Region_x','Returned']).nunique()
region_returns['Returns'] = region_returns['Order.ID']
region_returns = region_returns['Returns']
```

```python
region_returns = region_returns.reset_index()
```

```python
region_returns = region_returns.drop('Returned',axis=1)
region_returns
```

```python
region_return_rate = region_orders.merge(region_returns, how='outer', on='Region_x')
region_return_rate['Return_Rate'] = region_return_rate['Returns'] / region_return_rate['Order.ID']
region_return_rate.sort_values(by = 'Return_Rate', ascending=False)
```

```python
cat_orders = df.groupby('Category').nunique()['Order.ID'].reset_index()
cat_returns = df.groupby(['Category','Returned']).nunique()
cat_returns['Returns'] = cat_returns['Order.ID']
cat_returns = cat_returns['Returns']
cat_returns = cat_returns.reset_index()
cat_returns = cat_returns.drop('Returned',axis=1)
cat_return_rate = cat_orders.merge(cat_returns, how='outer', on='Category')
cat_return_rate['Return_Rate'] = cat_return_rate['Returns'] / cat_return_rate['Order.ID']
cat_return_rate.sort_values(by = 'Return_Rate', ascending=False)
```

```python
subcat_orders = df.groupby('Sub.Category').nunique()['Order.ID'].reset_index()
subcat_returns = df.groupby(['Sub.Category','Returned']).nunique()
subcat_returns['Returns'] = subcat_returns['Order.ID']
subcat_returns = subcat_returns['Returns']
subcat_returns = subcat_returns.reset_index()
subcat_returns = subcat_returns.drop('Returned',axis=1)
subcat_return_rate = subcat_orders.merge(subcat_returns, how='outer', on='Sub.Category')
subcat_return_rate['Return_Rate'] = subcat_return_rate['Returns'] / subcat_return_rate['Order.ID']
subcat_return_rate.sort_values(by = 'Return_Rate', ascending=False)
```
